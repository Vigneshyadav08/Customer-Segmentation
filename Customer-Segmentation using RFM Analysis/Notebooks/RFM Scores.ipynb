```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
retail_books = pd.read_csv('../Data/retail_RFM.csv')
```

```python
retail_books.head()
```

```python
retail_books.describe()
```

```python
plt.hist(retail_books['Monetary'], bins=65)
plt.title('Distribution of Monetary Values',fontname="Times New Roman",fontweight="bold")
plt.xlabel('Monetary Value',fontname="Times New Roman")
plt.ylabel('Frequency',fontname="Times New Roman")
plt.show()
```

```python
def m_score(x):
    if x >= 300:
        return 5
    elif x >= 200:
        return 4
    elif x >= 150:
        return 3
    elif x >= 100:
        return 2
    else:
        return 1

retail_books["M_Score"] = retail_books["Monetary"].apply(m_score)
```

```python
plt.hist(retail_books['Frequency'], bins=65)
plt.title('Distribution of Frequency Values',fontname="Times New Roman",fontweight="bold")
plt.xlabel('Frequency Value',fontname="Times New Roman")
plt.ylabel('Frequency',fontname="Times New Roman")
plt.show()
```

```python
def f_score(x):
    if x >= 3.5:
        return 5
    elif x >= 2:
        return 4
    elif x >= 2.5:
        return 3
    elif x >= 1.5:
        return 2
    else:
        return 1

retail_books["F_Score"] = retail_books["Frequency"].apply(f_score)
```

```python
plt.hist(retail_books['Recency'], bins=65)
plt.title('Distribution of Recency Values',fontname="Times New Roman",fontweight="bold")
plt.xlabel('Recency Value',fontname="Times New Roman")
plt.ylabel('Frequency',fontname="Times New Roman")
plt.show()
```

```python
def r_score(x):
    if x <= 30:
        return 5
    elif x <= 60:
        return 4
    elif x <= 90:
        return 3
    elif x <= 180:
        return 2
    else:
        return 1

retail_books["R_Score"] = retail_books["Recency"].apply(r_score)
retail_books.head()
```

```python
retail_books.to_csv('../Data/retail_RFM_Scores.csv', index=False)
```
